# KNeighborsClassifier(최근접 이웃)

```python
from sklearn.neighbors import KNeighborsClassifier                                 # (2, 2) 형태의 데이터를 받았으니 이것을 그대로 사용하고, 타깃 값을 정해주어 길이와 무게는 타깃이라는 답을 가진다.

kn = KNeighborsClassifier()
kn.fit(train_input, train_target)
kn.score(test_input, test_target)
```

```python
print(kn.predict([[25,150]]))
```

```python
import matplotlib.pyplot as plt

plt.scatter(train_input[:,0], train_input[:,1])
plt.scatter(25, 150, marker='^')
plt.grid()
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

```python
distances, indexes = kn.kneighbors([[25,150]])                                     # ()안에 들어간 좌표를 기준으로 가까운 자료를 위에서 설정한 n_neighbors의 개수만큼 보여준다.
```

```python
plt.scatter(train_input[:,0], train_input[:,1])
plt.scatter(25, 150, marker='^')
plt.scatter(train_input[indexes,0], train_input[indexes,1],
            marker='D', color='magenta')
plt.grid()
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

```python
print(train_input[indexes])
```

```python
print(train_target[indexes])
```

```python
print(distances)
```

```python
plt.scatter(train_input[:,0], train_input[:,1])
plt.scatter(25, 150, marker='^')
plt.scatter(train_input[indexes,0], train_input[indexes,1],
            marker='D', color='magenta')
plt.grid(axis='y')
plt.xlim((0, 1000))                                                                # x축과 y축의 비율이 같아야 그래프 상의 거리가 같아보이게 만들어진다.
plt.xlabel('langth')
plt.ylabel('weight')
plt.show()
```

```python
mean = np.mean(train_input, axis=0)
std = np.std(train_input, axis=0)
```

```python
print(mean, std)
```

```python
train_scaled = (train_input - mean) / std                                           # 표준편차
```

```python
new = ([25, 150] - mean) / std

plt.scatter(train_scaled[:,0], train_scaled[:,1])                                  # 0값이 그래프의 가운데로 왔다.
plt.scatter(new[0], new[1], marker='^')
plt.grid()
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```
